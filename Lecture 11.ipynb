### **Chapter 02. seaborn 라이브러리**

```python
import pandas as pd

df = pd.read_excel("(Dataset) Lecture 11 - 건강검진 데이터/health_screenings.xlsx")
df.head(3)
```

```python
df.columns
```

```python
df.info()
```

```python
df[['gender', 'smoking', 'drinking']].head(3)
```

```python
df['gender'] = df['gender'].astype(str) # int64 --> str
df['smoking'] = df['smoking'].astype(str)
df['drinking'] = df['drinking'].astype(str)

df.info()
```

```python
# DataFrame.loc[행, 열]
df.loc[df['gender'] == '1', ['gender']] = "Man"

df.loc[df['gender'] == '2', ['gender']] = "Woman"

df.gender.head(3)
```

```python
# smoking
df.loc[(df['smoking'] == '1') | (df['smoking'] == '2'), ['smoking']] = "Non-smoking"

df.loc[df['smoking'] == '3', ['smoking']] = "Smoking"

df.smoking.head(8)
```

```python
# drinking
df.loc[df['drinking'] == '0', ['drinking']] = "Non-drinking"

df.loc[df['drinking'] == '1', ['drinking']] = "Drinking"

df.drinking.tail(7)
```

```python
df[['gender', 'smoking', 'drinking']].head(3)
```

```python
df.columns
```

```python
smoking = df.groupby(["gender", "smoking"]).count()
smoking
```

```python
smoking = df.groupby(["gender", "smoking"])["smoking"].count()
smoking
```

```python
# Series --> to_frame() --> DataFrame
smoking = smoking.to_frame(name="count")
smoking
```

```python
smoking = smoking.reset_index()
smoking
```

```python
drinking = df.groupby(["gender", "drinking"])["drinking"].count()
drinking
```

```python
# Series --> to_frame() --> DataFrame
drinking = drinking.to_frame(name="count")
drinking
```

```python
drinking = drinking.reset_index()
drinking
```

```python
# matplotlib vs. seaborn --> bar() vs. baxplot()
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20, 5))

# superscript --> 지수
# subscript
fig.suptitle("matplotlib - bar graph", fontweight="bold")

index = np.arange(0, 4, 1)

fig.add_subplot(1, 2, 1) # (몇 개의 행, 몇 개의 열, 인덱스(1부터 카운팅됨))
plt.bar(index, smoking["count"])
plt.title("Smoking")
plt.ylabel("Count")
plt.xticks(index, ["Man Non-smoking", "Man Smoking", "Woman Non-smoking", "Woman Smoking"])

fig.add_subplot(1, 2, 2)
plt.bar(index, drinking["count"])
plt.title("Drinking")
plt.ylabel("Count")
plt.xticks(index, ["Man Drinking", "Man Non-drinking", "Woman Drinking", "Woman Non-drinking"])

plt.show()
```

```python
# matplotlib vs. seaborn --> bar() vs. baxplot()
import seaborn as sns

fig = plt.figure(figsize=(20, 5))

fig.suptitle("seaborn - bar graph", fontweight="bold")

index = np.arange(0, 4, 1)

area1 = fig.add_subplot(1, 2, 1) # (몇 개의 행, 몇 개의 열, 인덱스(1부터 카운팅됨))
area2 = fig.add_subplot(1, 2, 2)

sns.barplot(data=smoking, x="gender", y="count",
            hue="smoking", palette=["red", "limegreen"],
            ax=area1)
sns.barplot(data=drinking, x="gender", y="count",
            hue="drinking", palette=["lightblue", "gold"],
            ax=area2)

area1.set_title("Smoking")
area2.set_title("Drinking")

plt.show()
```

```python
# 산점도: scatter plot
# matplotlib vs. seaborn - scatter() vs. scatterplot()

df.head(3)
```

```python
df.columns
```

```python
m_data = df.loc[df['gender'] == "Man", ['gender', 'weight', 'waist']]
w_data = df.loc[df['gender'] == "Woman", ['gender', 'weight', 'waist']]
```

```python
m_data.head(3)
```

```python
w_data.head(3)
```

```python
fig = plt.figure()

plt.title("matplotlib - scatter plot")

plt.scatter(m_data["waist"], m_data["weight"], label="Man", c="blue", alpha=0.5)
plt.scatter(w_data["waist"], w_data["weight"], label="Woman", c="red", alpha=0.5)

plt.legend()
plt.xlabel("Waist")
plt.ylabel("Weight")
plt.grid()

plt.show()
```

```python
fig = plt.figure()

plt.title("seaborn - scatter plot")

sns.scatterplot(data=df, x="waist", y="weight", 
                hue="gender", palette=["blue", "red"])

plt.grid()

plt.show()
```

```python
# 히스토그램 (Histogram)
# hist() vs. histplot()

fig = plt.figure()

plt.title("matplotlib - Histogram")

plt.hist(m_data["weight"], bins=7, label="Man", 
         color="blue", alpha=0.5)
plt.hist(w_data["weight"], bins=7, label="Woman", 
         color="red", alpha=0.5)

plt.legend()
plt.xlabel("Weight")
plt.ylabel("Count")
plt.grid()

plt.show()
```

```python
# 히스토그램 (Histogram)
# hist() vs. histplot()

fig = plt.figure()

plt.title("matplotlib - Histogram")

plt.hist(m_data["weight"], bins=7, label="Man", 
         color="blue", alpha=0.5, edgecolor="gold")
plt.hist(w_data["weight"], bins=7, label="Woman", 
         color="red", alpha=0.5, edgecolor="black")

plt.legend()
plt.xlabel("Weight")
plt.ylabel("Count")
plt.grid()

plt.show()
```

```python
fig = plt.figure()

plt.title("seaborn - Histogram")

sns.histplot(data=m_data["weight"], bins=7, alpha=0.5, label="Man", color="blue")
sns.histplot(data=w_data["weight"], bins=7, alpha=0.5, label="Woman", color="red")

plt.grid()

plt.show()
```

```python
fig = plt.figure()

plt.title("seaborn - Histogram")

# 커널 밀도 추정치 (Kernel Density Estimate --> kde)
sns.histplot(data=m_data["weight"], bins=7, alpha=0.5, label="Man", color="blue",
             kde=True)
sns.histplot(data=w_data["weight"], bins=7, alpha=0.5, label="Woman", color="red",
             kde=True)

plt.grid()

plt.show()
```

```python
m_non_drink = df.loc[(df["drinking"] == "Non-drinking") & (df["gender"] == "Man"), "weight"]

w_non_drink = df.loc[(df["drinking"] == "Non-drinking") & (df["gender"] == "Woman"), "weight"]

m_drink = df.loc[(df["drinking"] == "Drinking") & (df["gender"] == "Man"), "weight"]

w_drink = df.loc[(df["drinking"] == "Drinking") & (df["gender"] == "Woman"), "weight"]
```

```python
# 상자 그림 --> boxplot()
fig = plt.figure()

plt.title("matplotlib - Box Plot")

plt.boxplot([m_non_drink, w_non_drink, m_drink, w_drink],
            tick_labels=["Man Non-drinking", "Woman Non-drinking", "Man Drinking", "Woman Drinking"])

plt.xlabel("drinking")
plt.ylabel("weight")
plt.grid()

plt.show()
```

```python
# 상자 그림 --> boxplot()
fig = plt.figure()

plt.title("matplotlib - Box Plot")

plt.boxplot([m_non_drink, w_non_drink, m_drink, w_drink],
            tick_labels=["Man Non-drinking", "Woman Non-drinking", "Man Drinking", "Woman Drinking"],
            vert=False) # 수직의 --> vertical

plt.ylabel("drinking")
plt.xlabel("weight")
plt.grid()

plt.show()
```

```python
# 상자 그림 --> boxplot()
fig = plt.figure()

plt.title("seaborn - Box Plot")

sns.boxplot(data=df, x="drinking", y="weight",
            hue="gender", palette=["blue", "red"])

plt.grid()

plt.show()
```

```python
# 상자 그림 --> boxplot()
fig = plt.figure()

plt.title("seaborn - Box Plot")

sns.boxplot(data=df, x="weight", y="drinking",
            hue="gender", palette=["blue", "red"],
            orient="h") # vertical -> "v" / horizontal -> "h"

plt.grid()

plt.show()
```

```python
# 카운트 플롯(Count Plot)

fig = plt.figure()

plt.title("seaborn - Count Plot")

sns.countplot(data=df, x="drinking", hue="gender", palette=["blue", "red"])

plt.grid()

plt.show()
```

```python
fig = plt.figure()

plt.title("seaborn - Count Plot")

sns.countplot(data=df, x="smoking", hue="gender", palette=["blue", "red"])

plt.grid()

plt.show()
```

```python
fig = plt.figure()

plt.title("seaborn - Count Plot")

sns.countplot(data=df, x="smoking", hue="gender", palette=["blue", "red"],
              order=["Smoking", "Non-smoking"])

plt.grid()

plt.show()
```

```python
fig = plt.figure()

plt.title("seaborn - Count Plot")

sns.countplot(data=df, y="smoking", hue="gender", palette=["blue", "red"],
              order=["Smoking", "Non-smoking"])

plt.grid()

plt.show()
```
