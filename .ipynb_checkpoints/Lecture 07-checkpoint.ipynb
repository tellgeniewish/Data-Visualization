### **Chapter 01. 숫자 데이터를 쉽게 다루도록 돕는 numpy 라이브러리**

```python
import matplotlib.pyplot as plt

a, b, c = [], [], []

for j in range(0, 11, 1):
    a.append(j)
    b.append(j ** 2)
    c.append(j ** 3)

plt.figure()
plt.plot(a, label='a', color="red")
plt.plot(b, label='b', color="green")
plt.plot(c, label='c', color="blue")

plt.grid()
plt.legend()
plt.show()
```

```python
for j in range(0, 1.1, 0.1): # range는 정수값을 입력받길 기대함
    a.append(j)
    b.append(j ** 2)
    c.append(j ** 3)
```

```python
for j in range(0, 11, 1): # 정수값을 기대함
    a.append(j / 10)
    b.append((j / 10) ** 2)
    c.append((j / 10) ** 3)
```

```python
a = range(0, 11, 1) ** 2
```

```python
import numpy as np

a = np.arange(0, 11, 1)
b = np.arange(0, 11, 1) ** 2 # 브로드캐스팅
c = np.arange(0, 11, 1) ** 3 # 브로드캐스팅

plt.figure()
plt.plot(a, label='a', color="red")
plt.plot(b, label='b', color="green")
plt.plot(c, label='c', color="blue")

plt.grid()
plt.legend()
plt.show()
```

```python
# import random
# random.randint() (random + integer)

# 표준정규분포: 평균이 0이고 표준편차가 1인 정규분포

# np.random.rand(m, n) # 균일분포에 따라서 0 ~ 1 사이의 실수값을 m-by-n 반환함!
# np.random.randn(m, n) # 표준정규분포에 따라서 m-by-n 반환함!
# np.random.randint(a, b) # 시작 숫자, 끝 숫자

a = np.random.rand(4, 3)
print(a)

b = np.random.randn(4, 3)
print(b)

c = np.random.randint(4, 21)
print(c)
```

```python
# np.random.choice() # 복원 추출 & 비복원 추출

# 복원 추출
print(np.random.choice(6, 11)) # (0 ~ 6 사이 11번 뽑아라) 리스트, 배열, 스칼라 값

# 비복원 추출
print(np.random.choice(6, 3, replace=False))
```

```python
# 확률 --> probability --> p
print(np.random.choice(6, 11, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1])) # 각각이 뽑힐 확률
```

```python
a = np.random.choice(6, 100000, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1])

plt.figure()
plt.hist(a, bins=6)
plt.show()
```

### **Chapter 02. numpy를 활용한 나만의 프로젝트 만들기**

```python
import csv

f = open("(Dataset) Lecture 05 - 인구 공공데이터/age.csv", 'r', encoding="cp949")
data = csv.reader(f)

header = next(data)

data = list(data)

name = input("인구 구조가 관심있는 동네 이름을 입력하세요: ")

# 내가 관심있는 동네를 찾는 부분!
# home = []
for row in data:
    # print(row)

    for j in range(2, 104, 1):
        row[j] = int(row[j].replace(',', ''))
        
    if name in row[0]:
        # for j in row[3:]:
            # home.append(int(j.replace(',', '')))
        
    # for j in range(2, 104, 1):
    #     row[j] = int(row[j].replace(',', ''))
            
        home = np.array(row[3:]) / row[2] # 인구 비율

# 비슷한 동네를 찾는 부분!
min_val = 10
result_name = ''
result = 0
for row in data:
    # print(row)
    # break
    
    # for j in range(2, 104, 1):
    #     row[j] = int(row[j].replace(',', ''))
            
    if (name not in row[0]) and (row[2] != 0):
        away = np.array(row[3:]) / row[2]

        s = np.sum((home - away) ** 2)
        if s < min_val:
            min_val = s
            result_name = row[0]
            result = away
        # break

f.close()

plt.rc("font", family="Malgun Gothic")

plt.figure()
plt.plot(home, label=name, color="blue")
plt.plot(result, label=result_name, color="red")
plt.xlabel("나이")
plt.ylabel("인구 비율")
plt.grid()
plt.legend()
plt.show()
```
