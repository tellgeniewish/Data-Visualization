### **Chapter 02. 데이터 가공**

```python
import pandas as pd

happy_life = pd.read_excel("(Dataset) Lecture 13 - 행복지수 데이터/대한민국행복지도_삶의만족도.xlsx")
happy_life.head(3)
```

```python
happy_safe = pd.read_excel("(Dataset) Lecture 13 - 행복지수 데이터/대한민국행복지도_안전.xlsx")
happy_safe.head(3)
```

```python
happy_environ = pd.read_excel("(Dataset) Lecture 13 - 행복지수 데이터/대한민국행복지도_환경.xlsx")
happy_environ.head(3)
```

```python
happy_econo = pd.read_excel("(Dataset) Lecture 13 - 행복지수 데이터/대한민국행복지도_경제.xlsx")
happy_econo.head(3)
```

```python
happy_edu = pd.read_excel("(Dataset) Lecture 13 - 행복지수 데이터/대한민국행복지도_교육.xlsx")
happy_edu.head(3)
```

```python
happy_relation = pd.read_excel("(Dataset) Lecture 13 - 행복지수 데이터/대한민국행복지도_관계및사회참여.xlsx")
happy_relation.head(3)
```

```python
happy_leisure = pd.read_excel("(Dataset) Lecture 13 - 행복지수 데이터/대한민국행복지도_여가.xlsx")
happy_leisure.head(3)
```

```python
happy_health = pd.read_excel("(Dataset) Lecture 13 - 행복지수 데이터/대한민국행복지도_건강.xlsx")
happy_health.head(3)
```

```python
# happy_life
# happy_safe
# happy_environ
# happy_econo
# happy_edu
# happy_relation
# happy_leisure
# happy_health
city = list(happy_life["시도"].unique())

happy_merge = pd.DataFrame({"시도": city}) # {key: value}
happy_merge
```

```python
# happy_life
life = happy_life["삶의 만족도"].groupby(by=happy_life["시도"]).mean()

happy_merge = pd.merge(happy_merge, life, on="시도")
happy_merge

# happy_safe
# happy_environ
# happy_econo
# happy_edu
# happy_relation
# happy_leisure
# happy_health
```

```python
# happy_life
# happy_safe
safe = happy_safe["평균"].groupby(by=happy_safe["시도"]).mean()

happy_merge = pd.merge(happy_merge, safe, on="시도")

happy_merge.rename(columns={"평균": "안전"}, inplace=True)

happy_merge

# happy_environ
# happy_econo
# happy_edu
# happy_relation
# happy_leisure
# happy_health
```

```python
# happy_life
# happy_safe
# happy_environ
environ = happy_environ["평균"].groupby(by=happy_environ["시도"]).mean()

happy_merge = pd.merge(happy_merge, environ, on="시도")

happy_merge.rename(columns={"평균": "환경"}, inplace=True)

happy_merge

# happy_econo
# happy_edu
# happy_relation
# happy_leisure
# happy_health
```

```python
# happy_life
# happy_safe
# happy_environ
# happy_econo
# happy_edu
# happy_relation
# happy_leisure
# happy_health
health = happy_health["평균"].groupby(by=happy_health["시도"]).mean()

happy_merge = pd.merge(happy_merge, health, on="시도")

happy_merge.rename(columns={"평균": "건강"}, inplace=True)

happy_merge
```

```python
# happy_life
# happy_safe
# happy_environ
# happy_econo
econo = happy_econo["평균"].groupby(by=happy_econo["시도"]).mean()

happy_merge = pd.merge(happy_merge, econo, on="시도")

happy_merge.rename(columns={"평균": "경제"}, inplace=True)

happy_merge

# happy_edu
# happy_relation
# happy_leisure
# happy_health
```

```python
# happy_life
# happy_safe
# happy_environ
# happy_econo
# happy_edu
edu = happy_edu["평균"].groupby(by=happy_edu["시도"]).mean()

happy_merge = pd.merge(happy_merge, edu, on="시도")

happy_merge.rename(columns={"평균": "교육"}, inplace=True)

happy_merge

# happy_relation
# happy_leisure
# happy_health
```

```python
# happy_life
# happy_safe
# happy_environ
# happy_econo
# happy_edu
# happy_relation
relation = happy_relation["평균"].groupby(by=happy_relation["시도"]).mean()

happy_merge = pd.merge(happy_merge, relation, on="시도")

happy_merge.rename(columns={"평균": "관계"}, inplace=True)

happy_merge

# happy_leisure
# happy_health
```

```python
# happy_life
# happy_safe
# happy_environ
# happy_econo
# happy_edu
# happy_relation
# happy_leisure
leisure = happy_leisure["평균"].groupby(by=happy_leisure["시도"]).mean()

happy_merge = pd.merge(happy_merge, leisure, on="시도")

happy_merge.rename(columns={"평균": "여가"}, inplace=True)

happy_merge

# happy_health
```
