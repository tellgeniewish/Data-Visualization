### **Chapter 01. 데이터 처리가 쉬운 판다스**

```python
import pandas as pd
import numpy as np

a = [80, 85, 90, 95, 100]

print(a)
print(type(a), "\n")

b = np.array(a)
print(b)
print(type(b), "\n")

c = pd.Series(a)
print(c)
print(type(c))
```

```python
d = pd.Series([11.11, 5.12, 3.14, 1.1415])
d
```

```python
a = [80, 85, 90, 95, 100]

e = pd.Series(a, index=["student1", "student2", "student3", "student4", "student5"])
print(e)
```

```python
import numpy as np

f = np.arange(0, 9, 1) # (start value, end value, step size)
print(f, "\n")

g = f.reshape(3, 3) # 3-by-3 matrix
print(g)
print(type(g), "\n")

h = pd.DataFrame(g)
print(type(h))
h
```

### **Chapter 03. 타이타닉 데이터 분석**

```python
titanic = pd.read_csv("(Dataset) Lecture 10 - 붓꽃 & 타이타닉 데이터/titanic.csv")
titanic
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic["Survived"]
```

```python
print(titanic.index)
print(titanic.columns)
```

```python
titanic["Survived"].unique()
```

```python
titanic["Survived"].value_counts()
```

```python
titanic.isna().sum()
```

```python
titanic.drop(["Cabin"], axis=1, inplace=True)
titanic
```

```python
titanic.isna().sum()
```

```python
titanic["Embarked"]
```

```python
titanic["Embarked"].value_counts()
```

```python
titanic["Embarked"] = titanic["Embarked"].fillna(titanic["Embarked"].mode())
titanic.isna().sum()
```

```python
print(titanic["Embarked"].mode())
print(type(titanic["Embarked"].mode()))  # Series
print(titanic["Embarked"].mode()[0])     # S
```

```python
titanic["Embarked"] = titanic["Embarked"].fillna('S')
titanic.isna().sum()
```

```python
titanic["Age"].mean()
```

```python
avg = titanic["Age"].mean()

titanic["Age"] = titanic["Age"].fillna(avg)

titanic.isna().sum()
```

```python
titanic
```

```python
titanic.columns
```

```python
titanic.drop(['Name', 'Ticket', 'Embarked'], axis=1, inplace=True)
titanic
```

```python
titanic.columns
```

```python
titanic.groupby(["Pclass", "Sex"]).mean()
```

```python
import seaborn as sns

sns.countplot(data=titanic, x="Pclass", hue="Survived",
              palette=["red", "green"],
              stat="count") # statistics: n. 통계, 통계학

plt.grid()
plt.show()
```

```python
sns.countplot(data=titanic, x="Pclass", hue="Survived",
             palette=["red", "green"],
             stat="percent") # statistics: n. 통계, 통계학

plt.grid()
plt.show()
```

```python
titanic.drop(["Sex"], axis=1, inplace=True)
titanic
```

```python
# 상관 분석(Correlation Analysis)
titanic.corr() # --> Sex가 숫자가 아니라서 날리고 해야 함
```

```python
sns.heatmap(data=titanic.corr())
plt.show()
```

```python
sns.heatmap(data=titanic.corr(), annot=True) # annotaion: n. 주석
plt.show()
```

```python
# annotaion: n. 주석
# color map --> cmap
sns.heatmap(data=titanic.corr(), annot=True, cmap="Blues")
plt.show()
```
