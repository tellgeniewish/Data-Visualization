### **Chapter 01. 데이터 처리가 쉬운 판다스**

```python
import pandas as pd
import numpy as np

a = [80, 85, 90, 95, 100]

print(a)
print(type(a), "\n")

b = np.array(a)
print(b)
print(type(b), "\n")

c = pd.Series(a)
print(c)
print(type(c))
```

```python
d = pd.Series([11.11, 5.12, 3.14, 1.1415])
d
```

```python
a = [80, 85, 90, 95, 100]

e = pd.Series(a, index=["student1", "student2", "student3", "student4", "student5"])
print(e)
```

```python
import numpy as np

f = np.arange(0, 9, 1) # (start value, end value, step size)
print(f, "\n")

g = f.reshape(3, 3) # 3-by-3 matrix
print(g)
print(type(g), "\n")

h = pd.DataFrame(g)
print(type(h))
h
```

```python
a = {"name": ["이상해", "파이리", "꼬부기"], # 키(Key): 값(Value)
    "number": [1, 4, 7],
     "type": ["grass", "fire", "water"]}

print(type(a))
print(a)

b = pd.DataFrame(a)
b
```
