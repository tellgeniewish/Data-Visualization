### **Lecture 01. 기본 그래프 그리기**

```python
import matplotlib.pyplot as plt

# 꺾은 선 그래프 그리기! --> plot([x1, x2, ...], [y1, y2, ...])
plt.figure()
plt.title("My First Graph")

plt.plot([1, 2, 3, 4], [3, 2, 4, 1], label="1st", # x, y의 개수가 일치하지 않으면 에러 발생
         color="r", linestyle=':', marker='s', linewidth=1.5)
plt.plot([1, 2, 3, 4], [1, 3, 2, 4], label="2nd", 
         c="gold", ls='-.', marker='o', lw=3.0) # o, s(정사각형), d(마름모), ^, v, >, < 등의 마커

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend() # n. 전설, 범례

plt.grid(True) # True는 생략 가능
plt.show()
```

### **Lecture 02. 크리스마스의 기온 변화를 그래프로 그리기**

```python
import csv

f = open("(Dataset) Lecture 03 & 04 - 기온 공공데이터/seoul.csv", "r", encoding="cp949")
data = csv.reader(f)

h = next(data)

avg_temp, min_temp, max_temp = [], [], []
for row in data:
    if row[4] != '':
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '25':
            avg_temp.append(float(row[2]))
            min_temp.append(float(row[3]))
            max_temp.append(float(row[4]))

f.close()

print(len(result)) # len -> length
print("Before: ", result)
```

```python
# 한글을 사용하고 싶을 때
plt.rc("font", family="Malgun Gothic") # Window OS
# plt.rc("font", family="AppleGothic") # MacOS

plt.rcParams["axes.unicode_minus"] = False # 마이너스 기호에 대한 경고 사라짐

plt.figure(figsize=(10, 2))

plt.title("크리스마스 기온 변화 추이")

plt.plot(avg_temp, c='g', label="Avg Temperature")
plt.plot(min_temp, c='b', label="Min Temperature")
plt.plot(max_temp, c='r', label="Max Temperature")

plt.xlabel("Day")
plt.ylabel("Temperature")

plt.legend()

plt.grid(True)
plt.show()
```

```python
# 지구 온난화 확인하기!
# 연평균 기온이 상승하고 있다!

import csv

f = open("(Dataset) Lecture 03 & 04 - 기온 공공데이터/seoul.csv", "r", encoding="cp949")
data = csv.reader(f)

h = next(data)

year = [] # 꺾은 선 그래프의 x축(가로축)의 좌표값을 지정하기 위한 용도!
day = [] # 365일치 기온 값을 지정하기 위한 용도!
mean = [] # 연평균 기온 값을 저장! --> 꺾은 선 그래프의 y축의 좌표값으로 활용!

for row in data:
    y = int(row[0].split('-')[0]) # str --> int
    if y not in [1907, 1950, 1951, 1952, 1953, 2024]:
        if row[2] != '':
            day.append(float(row[2])) # str --> float
            if row[0].split('-')[1] == "12" and row[0].split('-')[2] == "31":
                mean.append(sum(day)/len(day))
                year.append(y)
                
                day = [] # 초기화!
                
    elif y == 2024:
        break

f.close()
```

```python
plt.figure(figsize=(10, 2))
plt.title("서울의 연평균 기온 변화 추이")

plt.plot(year, mean, c='y', marker='^')

plt.xlabel("Year")
plt.ylabel("Avg Temperature")

plt.grid()
plt.show()
```

### **Chapter 03. 기온 데이터를 다양하게 시각화하기**

```python
# 1) 히스토그램 (Histogram) -> hist
import matplotlib.pyplot as plt

plt.figure()

plt.hist([1, 1, 1, 2, 2, 3, 3, 3, 4, 5, 5, 5, 5, 5])

plt.show()
```

```python
# 주사위 던지는 행위를 반복해서, 주사위의 눈이 나오는 횟수를 시각화하기!
import random

# print(random.randint(1, 6)) # 시작 숫자, 끝 숫자(미만 아님! 포함임!)

d = []
for _ in range(0, 100, 1): # 변수 만들 필요 없으면 _를 사용하면 됨
    d.append(random.randint(1, 6))

print(d)

# 경험적 확률 --> 시행에 따라서, 계산되는 확률 (표본)
# 수학적 확률 --> 1/6 (모집단)
# 큰 수의 법칙 (Low of Larger Number, LLN): 시행 횟수가 늘어나면 모집단의 분포를 따라간다

plt.figure()

plt.hist(d, bins=6, color="red") # bin <-- n. 깡통

plt.xlabel("Dice")
plt.ylabel("Count")

plt.grid()

plt.show()
```

```python
f = open("(Dataset) Lecture 03 & 04 - 기온 공공데이터/seoul.csv", "r", encoding="cp949")
data = csv.reader(f)

header = next(data)

jan = []
aug = [] # august <-- 8월
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            aug.append(float(row[-1])) # str -> float
        elif row[0].split('-')[1] == '01':
            jan.append(float(row[-1])) # str -> float

f.close()

plt.figure()
plt.hist(aug, bins=100, color="red", label="August")
plt.hist(jan, bins=100, color="blue", label="January")

plt.xlabel("Temparature")
plt.ylabel("Count")
plt.legend()

plt.grid()
plt.show()
```

```python
f = open("(Dataset) Lecture 03 & 04 - 기온 공공데이터/seoul.csv", "r", encoding="cp949")
data = csv.reader(f)

header = next(data)

day = []
for _ in range(0, 31, 1): # 3월
    day.append([])
    
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == "03":
            day[int(row[0].split('-')[2]) - 1].append(float(row[-1])) # str -> float

f.close()

plt.figure()
plt.boxplot(day)

plt.xlabel("Day")
plt.ylabel("Temparature")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot(day)

plt.xlabel("Day")
plt.ylabel("Temparature")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot(day, showfliers=False)

plt.xlabel("Day")
plt.ylabel("Temparature")
plt.grid()
plt.show()
```

```python
plt.style.use("ggplot")

plt.figure(figsize=(10, 5))
plt.boxplot(day, showfliers=False)

plt.xlabel("Day")
plt.ylabel("Temparature")
plt.grid()
plt.show()
```
