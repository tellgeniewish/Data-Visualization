### **Lecture 03. 서울이 가장 더웠던 날은 언제였을까?**

```python
import csv

f = open("(Dataset) Lecture 03 & 04 - 기온 공공데이터/seoul.csv", "r", encoding="cp949") # 윈도우는 cp949 생략 가능, 리눅스나 맥은 명시해야 함
data = csv.reader(f)

h = next(data) # header

# for row in data:
#     print("첫 출력!", row)
#     break
    
max_date = ''
max_temp = -999

for row in data:
    # 결측지(Missing Value) 대처하는 방법
    # 1) 제거(Delete)
    # 2) 대체(Imputation)
    
    if row[4] == '':
        row[4] = -999
    else:
        row[4] = float(row[4]) # str -> float
    # print(row)

    if max_temp < row[4]:
    	max_temp = row[4]
    	max_date = row[0]
    
f.close()
```

```python
print("기상 관측 이래로 서울의 가장 더웠었던 날짜와 그때의 기온은?")
print("날짜:", max_date)
print("기온:", max_temp)
```
