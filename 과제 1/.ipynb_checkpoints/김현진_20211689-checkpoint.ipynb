<h3><span style="color:black; background-color:skyblue"><b>문제 1 (0.5 points)</b></span></h3>
폭염일수는 일 최고기온이 33℃ 이상인 날의 수를 의미합니다. heat_wave.csv 파일는 2014년도부터 2023년도까지 매년 폭염 지속일수가 가장 길었던 지점의 이름과 시작일, 종료일 데이터가 기록돼 있습니다. 기록된 데이터 중에서 폭염 지속일수가 가장 길었던 연도를 찾고, 지점, 시작일, 종료일을 출력시킬 수 있도록, 알고리즘을 구현하세요.

<h4><b>제약사항 및 주의사항</b></h4>
    
* 외부 라이브러리(예: pandas)에서 제공하는 함수 또는 클래스 등을 사용할 경우, 0점 처리됩니다.<br>
* 아래 '결과 화면'과 **100% 동일한** 결과가 출력되도록, print( ) 함수를 적절히 활용하세요.<br>
* 출력되는 결과 화면에서 띄어쓰기도 평가 요소에 포함되며, 동일하지 않은 정도에 따라 감점이 있습니다.<br>
* '결과 화면'에서 <span style="color:blue">파란색</span>으로 표시된 부분은, 문자열이 아닌 변수를 활용해 출력돼야 합니다.<br>

<h4><b>결과 화면</b></h4>
폭염 지속일수가 가장 길었던 연도: <span style="color:blue">2018</span>년<br>
해당 지점: <span style="color:blue">금산</span><br>
폭염 시작일: <span style="color:blue">07</span>월 <span style="color:blue">11</span>일<br>
폭염 종료일: <span style="color:blue">08</span>월 <span style="color:blue">16</span>일<br>
지속일수: <span style="color:blue">37</span>일

```python
import csv

f = open("heat_wave.csv", 'r', encoding="cp949")
data = csv.reader(f)

# 아래부터 솜솜이가 작성해야 할 부분
h = next(data)
h = next(data)
h = next(data)
h = next(data)
h = next(data)

last = -1
for row in data:
    if last < int(row[-1]):
        last = int(row[-1])
        year = int(row[0])
        loc = row[1]
        start = row[2]
        end = row[3]

print("폭염 지속일수가 가장 길었던 연도: %d년" % year)
print("해당 지점:", loc)
print(f"폭염 시작일: {start.split('-')[1]}월 {start.split('-')[2]}일")
print(f"폭염 종료일: {end.split('-')[1]}월 {end.split('-')[2]}일")
print("지속일수: %d일" % last)
    
# 수고하셨습니다.
f.close()
```

<h3><span style="color:black; background-color:skyblue"><b>문제 2 (0.5 points)</b></span></h3>
sr_profic.csv 파일에는 주식회사 SR의 연도별 SRT 운행 수익 데이터가 기록돼 있습니다. 2016년도부터 2023년까지 홀수 연도에 대해, 운행 노선 중 하나인 경부선 운행에 따른, 월별 SRT 운행 수익을 꺾은선 그래프로 시각화하세요.

<h4><b>제약사항</b></h4>

* 외부 라이브러리(예: pandas)에서 제공하는 함수 또는 클래스 등을 사용할 경우, 0점 처리됩니다.<br>
* 그래프의 가로 및 세로 길이는 기본값으로 사용하세요.<br>
* 그래프의 가로세로 축의 이름, 격자, 범례 등 세부 사항은 셀 아래 그래프를 참고하여 설정하세요.<br>
* 각 선 그래프의 선 굵기는 모두 0.5로 설정하세요.<br>
* 선 그래프의 색상은 연도 순으로 빨간색, 초록색, 파란색, 검은색으로 설정하세요.<br>
* 마커는 자유롭게 설정하되, 각 선 그래프마다 서로 다른 마커를 적용하세요.<br>

<h4><b>솜솜이가 얻어야 하는 그래프</b></h4>

```python
import csv
import matplotlib.pyplot as plt

f = open("sr_profit.csv", 'r', encoding="cp949")
data = csv.reader(f)

# 아래부터 솜솜이가 작성해야 할 부분
h = next(data)

months = list(range(1, 13))

srt_17 = []
srt_19 = []
srt_21 = []
srt_23 = []

for row in data:
    if "2017년 경부선" in h:
        srt_17.append(int(row[h.index("2017년 경부선")]))    
    if "2019년 경부선" in h:
        srt_19.append(int(row[h.index("2019년 경부선")]))
    if "2021년 경부선" in h:
        srt_21.append(int(row[h.index("2021년 경부선")]))
    if "2023년 경부선" in h:
        srt_23.append(int(row[h.index("2023년 경부선")]))

plt.figure()

plt.plot(months, srt_17, marker='^', c='r', lw=0.5, label="2017")
plt.plot(months, srt_19, marker='d', c='g', lw=0.5, label="2019")
plt.plot(months, srt_21, marker='o', c='b', lw=0.5, label="2021")
plt.plot(months, srt_23, marker='v', c='black', lw=0.5, label="2023")

plt.xlabel("Month")
plt.ylabel("Profit")

plt.legend()
plt.grid()

plt.show()

# 수고하셨습니다.
f.close()
```
