```python
dust = pd.read_excel("(Dataset) Lecture 10 - 붓꽃 & 타이타닉 데이터/dust.xlsx")
dust.head()
```

```python
dust.columns
```

```python
dust2 = dust.drop(['지역', '망', '측정소코드'], axis=1)
dust2.head()
```

```python
dust2.columns
```

```python
dust3 = dust.drop(['주소'], axis="columns")
dust3.head()
```

```python
dust3.columns
```

```python
dust2["city"] = "서울"
dust2.head()
```

```python
dust2.rename(columns={"측정소명": "name",
                     "측정일시": "date",
                     "주소": "addr"}, inplace=True)
dust2.columns
```

```python
dust2.dtypes
```

```python
dust2["date"] = dust2["date"].astype(str)
dust2["date"] = dust2["date"].str.slice(0, 8)
print(dust2["date"].dtypes, "\n")
print(dust2.head(), "\n")

dust2["date"] = pd.to_datetime(dust2["date"])
print(dust2["date"].dtypes)
dust2.head()
```
