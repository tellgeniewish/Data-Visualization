<h3><span style="color:black; background-color:skyblue"><b>문제 1 (0.5 points)</b></span></h3>
폭염일수는 일 최고기온이 33℃ 이상인 날의 수를 의미합니다. heat_wave.csv 파일는 2014년도부터 2023년도까지 매년 폭염 지속일수가 가장 길었던 지점의 이름과 시작일, 종료일 데이터가 기록돼 있습니다. 기록된 데이터 중에서 폭염 지속일수가 가장 길었던 연도를 찾고, 지점, 시작일, 종료일을 출력시킬 수 있도록, 알고리즘을 구현하세요.

<h4><b>제약사항 및 주의사항</b></h4>
    
* 외부 라이브러리(예: pandas)에서 제공하는 함수 또는 클래스 등을 사용할 경우, 0점 처리됩니다.<br>
* 아래 '결과 화면'과 **100% 동일한** 결과가 출력되도록, print( ) 함수를 적절히 활용하세요.<br>
* 출력되는 결과 화면에서 띄어쓰기도 평가 요소에 포함되며, 동일하지 않은 정도에 따라 감점이 있습니다.<br>
* '결과 화면'에서 <span style="color:blue">파란색</span>으로 표시된 부분은, 문자열이 아닌 변수를 활용해 출력돼야 합니다.<br>

<h4><b>결과 화면</b></h4>
폭염 지속일수가 가장 길었던 연도: <span style="color:blue">2018</span>년<br>
해당 지점: <span style="color:blue">금산</span><br>
폭염 시작일: <span style="color:blue">07</span>월 <span style="color:blue">11</span>일<br>
폭염 종료일: <span style="color:blue">08</span>월 <span style="color:blue">16</span>일<br>
지속일수: <span style="color:blue">37</span>일

```python
import csv

f = open("heat_wave.csv", 'r', encoding="cp949")
data = csv.reader(f)

# 아래부터 솜솜이가 작성해야 할 부분
for _ in range(0, 4, 1):
    next(data)

h = next(data)
# print(h)

year = ''
pos = ''
start_date = ''
end_date = ''
days = ''

max_day = 0
for row in data:
    if max_day < int(row[4]):   # row[-1] 가능!
        max_day = int(row[4])
        
        year = row[0]
        pos = row[1]
        start_date = row[2]
        end_date = row[3]
        days = row[4]
    
    
    
print("폭염 지속일수가 가장 길었던 연도:", year + "년")
print("해당 지점:", pos)
print("폭염 시작일:", start_date.split('-')[1] + "월", 
                     start_date.split('-')[2] + "일")
print("폭염 종료일:", end_date.split('-')[1] + "월",
                     end_date.split('-')[2] + "일")
print("지속일수:", days + "일")
    
# 수고하셨습니다.
f.close()
```

<h4><b>솜솜이가 얻어야 하는 그래프</b></h4>

```python
import csv
import matplotlib.pyplot as plt

f = open("sr_profit.csv", 'r', encoding="cp949")
data = csv.reader(f)

# 아래부터 솜솜이가 작성해야 할 부분
h = next(data)

profit = []
for _ in range(7):
    profit.append([])
    
for row in data:
    for j in range(7):
        profit[j].append(int(row[3 * j + 4]))  # 4, 7, 10, 13, 16, 19, 22 -> 4 + (n-3)*3 = 3j + 4
        
plt.figure()
plt.plot(range(1, 13, 1), profit[0], marker='^', label="2017", lw=0.5, color="red")
plt.plot(range(1, 13, 1), profit[2], marker='>', label="2019", lw=0.5, color="green")
plt.plot(range(1, 13, 1), profit[4], marker='v', label="2021", lw=0.5, color="blue")
plt.plot(range(1, 13, 1), profit[6], marker='<', label="2023", lw=0.5, color="black")
plt.legend()
plt.xlabel("Month")
plt.ylabel("Profit")
plt.grid()
plt.show()

# 수고하셨습니다.
f.close()
```
